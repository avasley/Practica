```python
# También son conocidas como métodos

# Para crear un metodo utilizamos: def
# Tenemos 4 métodos:
# -Métodos Vacios
# -Métodos que reciben valores
# -Métodos que devuelven valores
# -Métodos que reciben y devuelven valores
```

```python
# Métodos Vacios
# Explicación: No reciben valores(Parámetros)

def saludo():
    print("Bienvenidos al curso de LP3")
    print("****Python****")
    saludo()
```

```python
saludo()
```

```python
# Mñetodo que reciben valores

def factorial(numero):
    fac = 1
    for i in range(1, numero+1):
        fac*=i
        #fac = fac*i
    print(f"el factorial es: {fac}")
```

```python
numero = int(input("ingrese Número: "))
factorial(numero)
```

```python
# Método que devuelven valores
def saludo2():
    mensaje = "Bienvenidos a LP3"
    return mensaje
```

```python
print(saludo2())
```

```python
# Método que recibe y devuelvem valores

def factorial2(numero):
    fac = 1 
    for i in range(1, numero+1):
        fac*=i
        # fac = fac*i
    return fac
```

```python
numero = int(input("Número: "))
print(f"El factorial de {numero} es {factorial2(numero)}")
```
